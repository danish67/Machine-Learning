```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(3)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Per capita income')
plt.scatter(df.year,df.percapinc)
```

```python
reg= linear_model.LinearRegression()
#reg.fit(df[[kis ke hisaab]],df.kis ke hisaab se kya)
reg.fit(df[['year']],df.percapinc)
```

```python
reg.predict([[2023]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Per capita income')
plt.scatter(df.year,df.percapinc)
plt.plot(df.year,reg.predict(df[['year']]),color='red')
```
